```python
# Calcular la distancia entre dos puntos
# Ingresar los valores de X1, Y1, X2, Y2
# Formula para calcular la distancia
# X = X2 - X1
# Y = Y2 - Y1
# D = SQRT(X**2 + Y**2)
# mostrar el valor de la distancia D
from math import sqrt as SQRT

X1 = int(input("Ingrese el valor de X1: "))
Y1 = int(input("Ingrese el valor de Y1: "))
X2 = int(input("Ingrese el valor de X2: "))
Y2 = int(input("Ingrese el valor de Y2: "))

X = X2 - X1
Y = Y2 - Y1

D = SQRT(X**2 + Y**2)

print("La distancia es: ", D)
print("Programa realizado por Karla Saenz = 48")
```
