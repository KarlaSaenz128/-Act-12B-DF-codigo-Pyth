```python
# calcula la distancia entre dos puntos
# ingresa los valores de X1, Y1, X2, Y2
# formula para calcular la distancia
# x = X2 - X1
# Y = Y2 - Y1
# D = SQRT(X**2 + Y**2)
# Mostrar el valor de la distancia
from math import sqrt
X1 = float(input("Ingrese el valor de X1: "))
Y1 = float(input("Ingrese el valor de Y1: "))
X2 = float(input("Ingrese el valor de X2: "))
Y2 = float(input("Ingrese el valor de Y2: "))
X = X2 - X1
Y = Y2 - Y1
D = sqrt(X**2 + Y**2)
print("La distancia entre dos pumtos es:", D)
print("programa realizado por Karla Saenz NL = 48")
```
